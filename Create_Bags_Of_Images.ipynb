```python
import os,sys
import h5py
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator,array_to_img, img_to_array, load_img
from keras.models import Sequential
from keras.layers import Dropout, Flatten, Dense
from keras import applications
import matplotlib.pyplot as plt
import seaborn as sns
import math
from tqdm import tqdm
from PIL import Image
data_root='.'
from random import randrange

#Create 5 different "bags of words"
for x in range(16,22): 
    j=x
    if x>19:
        fill='neares'
    else:
        fill='constant'
    
    #Define output path
    x=str(x)
    path = "C:\\Users\\anne.klaver\\Documents\\Image-classification-transfer-learning-master\\image_bag_"+x 

    try:
        os.mkdir(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
    else:
        print ("Successfully created the directory %s " % path)
    
    #Add 80 (1-80) empty folders to the output directory
    for x in range(1,81): 
        x=str(x)
        path2 = path+'\\'+x
        try:
            os.mkdir(path2)
        except OSError:
            print ("Creation of the directory %s failed" % path)
        else:
            print ("Successfully created the directory %s " % path)
    
    datagen = ImageDataGenerator(
            rotation_range=j*2.5,
            width_shift_range=j*0.018,
            height_shift_range=j*0.018,
            shear_range=j*0.018,
            zoom_range=j*0.018,
            horizontal_flip=True,
            fill_mode=fill)
    
    #location of original train data
    src_train_dir=os.path.join(data_root,'train_set_old\\train_set')
    dest_train_dir=os.path.join(path)
    
    #For each class, read the images, decide how many times to use them, 
    #and apply data augmentation before saving it to the corresponding folder
    for x in range (1,81): 
        dest_lab_dir=os.path.join(dest_train_dir,str(x))
        src_lab_dir=os.path.join(src_train_dir,str(x))
        for file in os.listdir(src_lab_dir):
                img=load_img(os.path.join(src_lab_dir,file))
                joe=img_to_array(img) 
                joe=joe.reshape((1,) + joe.shape)
                i=10
                
                #When class 20, save the picture 10??????????? times (due to the extreme small size of class 20)
                if x==20:
                    i=0
                
                #Do not pick the image with a probability of 1/4??????????????
                getal=randrange(7,11)
                
                if getal !=10:
                    for joe in datagen.flow(joe, batch_size=1,save_to_dir=dest_lab_dir, save_format='jpg'):
                        i+=1
                        if i>9: #save only one version of each image, decreate in case you want multiple variations
                            break
```
